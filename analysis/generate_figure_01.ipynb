# Generate Figure 1
Figure 1 shows the distributions of `measure_a` and `measure_b`.

```python
%matplotlib inline
import pandas as pd
import seaborn as sns

sns.set_style("whitegrid")
```

```python
# Read in the data
in_file = "../data/demographics.tsv"
df = pd.read_table(in_file)

# Generate the plot
plot = sns.jointplot(data=df, x="measure_a", y="measure_b", kind="reg", truncate=False)
plot.figure.savefig("../figures/figure_01.svg")
```
